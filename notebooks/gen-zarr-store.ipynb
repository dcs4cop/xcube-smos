```python
from functools import cached_property
from functools import lru_cache

import numpy as np
import xarray as xr

from xcube.core.zarrstore import GenericArray
from xcube.core.zarrstore import GenericZarrStore
```

```python
import dask.distributed
dask.distributed.Client(processes=True, n_workers=8)
```

```python
class Test:
    def __init__(self, size, chunks):
        self._size = size
        self._chunks = chunks
    
    def __getstate__(self):
        import sys
        state = self.__dict__.copy()
        #print(repr(state), file=sys.stderr, flush=True)
        state.pop("shape", None)
        state.pop("chunks", None)
        return state
    
    @property
    #@cached_property
    def shape(self):
        return self._size,

    @property
    #@cached_property
    def chunks(self):
        return self._chunks,

    #@lru_cache()
    def work_hard(self, todo):
        import time
        time.sleep(todo)
        return np.arange(0, self._chunks, dtype=np.float32)
```

```python
test = Test(size=10000, chunks=100)
```

```python
zarr_store = GenericZarrStore(
    GenericArray(name="x", dims="x",
                 data=np.arange(test.shape[0])),
    GenericArray(name="y", dims="x",
                 get_data=test.work_hard,
                 get_data_params={"todo": 1.5},
                 shape=test.shape,
                 chunks=test.chunks, 
                 dtype=np.dtype("float32"))
)
```

```python
ds = xr.open_zarr(zarr_store)
ds
```
